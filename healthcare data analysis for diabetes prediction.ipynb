# 2)examination of diabetic and non_deavetic people with other issues.

```python
var_cols = ['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI']
plot = 0
fig=plt.figure(figsize=(8,8))
for i in range(len(var_cols)):
    plot+=1
    ax=plt.subplot(3,2,plot)
    sns.boxplot(x=df["Outcome"],y=df[var_cols[i]],palette="viridis")
    plt.title("diavetic and non diavetic people with oher issues")
plt.tight_layout()
plt.show()
```
